# Load Data

```python
train = load_dataset("LabHc/bias_in_bios", split= "train")
test = load_dataset("LabHC/bias_in_bios", split= "test")
dev = load_dataset("LabHC/bias_in_bios", split= "dev")
```

```python
# Inspect structure
print('Train Structure')
print(train)
print()
print('Test Structure')
print(test)
print()
print('Dev Structure')
print(dev)
```

# EDA
(See Christine's notebook 01 for full EDA)

```python
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)
dev_df = pd.DataFrame(dev)
```

```python
# Check for NAs
print('Train NAs')
print(train_df.isna().sum())
print()
print('Test NAs')
print(test_df.isna().sum())
print()
print('Dev NAs')
print(dev_df.isna().sum())
```

# Save to drive

```python
# Update path direction to data folder
path = path + '/data_original'
```

```python
train.save_to_disk(path + '/train')
test.save_to_disk(path + '/test')
dev.save_to_disk(path + '/dev')
```
